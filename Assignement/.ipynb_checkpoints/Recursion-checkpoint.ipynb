```python
#Find Index of key present

def firstIndex(arr, x):
    # Please add your code here
    l=len(arr)
    if(l==0):
        return -1
    if(arr[0]==x):
        return 0
    flag=True


    smallOutput=firstIndex(arr[1:],x)
    if(smallOutput==-1):
        flag=False
        return -1
    if(flag):
        return 1+smallOutput
    else:
        return -1


def firstIndexBetter(arr,x,si):
    l=len(arr)
    if si==l:
        return -1
    if arr[si]==x:
        return si
    
    return firstIndexBetter(arr,x,si+1)
    
arr=[1,2,3,4,5,3,4,5,8]
x=9
print(firstIndex(arr,x))
print(firstIndexBetter(arr,x,0))


    
```

```python
def replaceChar(s,a,b):
    if(len(s)==0):
        return s
    smallOutput= replaceChar(s[1:],a,b)
    if(s[0]==a):
        return b+smallOutput
    else:
        return s[0]+smallOutput
    
```

```python
#Replace pi with 3.14

def replacePi(string):
    
    if len(string)<=1:
        return string
    
    if (string[0]=='p' and string[1]=='i'):
        smallOutput=replacePi(string[2:])
        return '3.14'+ smallOutput

    else:
        smallOutput=replacePi(string[1:])
        return string[0]+smallOutput



# Main
string = input()
print(replacePi(string))
```

```python
#Binary Search using Recursion

def binarySearch(si,ei,arr,key):
    
    if(si>ei):
        return -1
    
    mid=(si+ ei)//2
    if(arr[mid]==key):
        return mid
    elif (arr[mid]>key):
        return binarySearch(si,mid-1,arr,key)
    else:
        return binarySearch(mid+1,ei,arr,key)
    
arr=[1,4,9,10,13,15] //array should be sort
binarySearch(0,5,arr,13)
```

```python
from sys import setrecursionlimit
setrecursionlimit(11000)

def partition(arr,start, end):
    print("start is= " + str(start))
    print("end is= "+ str(end))
    
    k=arr[start]
    print("k is =" + str(k))
    count=0
    for ele in range(start,end):
        if(arr[ele] <k):
            print("ele is =" +str(arr[ele]))
            count=count+1
    if(count==0):
        return start
    #Swaping element
    temp=arr[start +count]
    arr[start+ count]=k
    arr[start]=temp
    print("count is" + str(count))

    i= start
    j= end-1
    print("i = " + str(i))
    print("j = " + str(j))
    while(i<=j and i!= count+start):
        if(arr[i]<arr[count+start]):
            i=i+1
            print("i = " + str(i))
        elif(arr[j]>=arr[count+start]):
            j=j-1
            print("j = " + str(j))
            print("array[count] =" + str(arr[count]))
        else:
            print(arr)
            print("Swaping " + str(arr[i]) + " and " + str(arr[j])  )
            temp= arr[i]
            arr[i]=arr[j]
            arr[j]=temp 
            print(arr)
            i=i+1
            j=j-1
            
    return start+count




def quickSort(arr, start, end):
    # Please add your code here
    if(start >= end):
        return
    p=partition(arr,start,end)
    quickSort(arr,start,p-1)
    quickSort(arr,p+1,end)

n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
quickSort(arr, 0, n)
print(*arr)

```

```python
#TowerOfHanoi

def tower_hanoi(n,source,aux, dest):
    if(n==1):
        print(source + " " + dest)
        return
    tower_hanoi(n-1, source,dest,aux)
    print(source + " " + dest)
    
    tower_hanoi(n-1, aux,source,dest)
    
tower_hanoi(4,'a','b','c')
```

```python
def correct_str(s,start,end):
    print(s[start:end])
    print(str(start))
    print(str(end))
    if(start==end):
        return True
    if(start==0 ):
        if (s[start]=='a'):
            if(len(s)==1):
                return True
        else:
            return False

    
    if(s[start]=='a' and s[start+1]=='a'):
        return correct_str(s, start+1, end)
    elif(s[start]=='a' and s[start+1]=='b' and s[start+2]=='b'):
        return correct_str(s, start+3, end)
    elif(s[start]=='b' and s[start+1]=='b' and start+2==end):
        return correct_str(s, start+2, end)
    elif(s[start]=='b' and s[start+1]=='b' and s[start+2]=='a'):
        return correct_str(s, start+3, end)
    else:
        return False





s=input()
l=len(s)
x=correct_str(s,0,l)
if(x):
    print("true")
else:
    print("false")
```

```python
#Does s contain t ?
def contains(s,t):
    #Implement This Function Here
    if(len(s)==0 and len(t)!=0):
        return False

    if(len(t)==0):
        return True

    for i in range(0,len(s)):
        if t[0]==s[i]:
            print("found t[0]")
            contains(s[i+1:],t[1:])

s = 'mmdhfdajdhdamiuhiuhasdgsgf'
t = 'mahlyfmahrulz'

ans = contains(s,t)
if ans is True:
    print('true')
else:
    print('false')
```
